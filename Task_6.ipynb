 - <b>Step 1:</b> Import all required libraries

```python
import sklearn.datasets as datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

 - <b>Step 2:</b> Load the iris dataset

```python
# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(10)
```

```python
y=iris.target
print(y)
print(iris.target_names)
```

The target array prints the response variables (label) of the data.<br>target_names gives the name(s) of the target variable(s).

 - <b>Step 3:</b> Define the Decision Tree Algorithm

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='entropy',min_samples_leaf=0.02,random_state=0)
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

The attributes to the DecisionTreeClassifier are:
<br>criterion: The function to measure the quality of a split. Information gain is the entropy of parent node minus sum of weighted entropies of child nodes. 
<br>min_samples_leaf: The minimum number of samples required to be at a leaf node.
<br>random_state: Controls the randomness of the estimator.

 - <b>Step 4:</b> Visualize the Decision Tree

```python
from sklearn.tree import plot_tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
plot_tree(dtree,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

 - <b>Step 5:</b> Model Evaluation

```python
from sklearn.model_selection import cross_val_score
print("mean:",cross_val_score(dtree,iris.data,iris.target, cv=5).mean())
print("standard deviation:",cross_val_score(dtree,iris.data,iris.target, cv=5).std())
```

cross_val_score computes the cross validation score on the dataset by splitting the data, fitting a model and computing the score 5 consecutive times (cv=5), with different splits each time. Then the mean and standard deviation of the scores of each CV iteration are taken to estimate the accuracy.

Any new/test data can be fed to this classifer and it would be able to predict the right class accordingly.
