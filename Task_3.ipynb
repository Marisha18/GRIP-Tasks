 - <b>Step 3:</b> Exploratory Data Analysis and Data Visualization

###### Heatmap of Correlation

```python
# Data Correlation
correlation = df.corr()
sns.heatmap(correlation, annot = True, cmap = 'magma')
```

From the above Heatmap we can conclude that:
<br>Sales and Profit are moderately correlated
<br>Discount and Profit are negatively correlated
<br>Quantity and Profit are less negatively correlated

###### Kernel Density Estimate Plot
Kernel Density Estimate(KDE) is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

```python
# KDEplot
plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'], color = 'red', label = 'Sales', shade = True, bw=20)
sns.kdeplot(df['Profit'], color = 'blue', label = 'Profit', shade = True, bw=20)
plt.legend()
```

Profit is more than that of sales, but there are some areas where profit could be increased.

###### Pair-Plots
Seaborn Pairplot uses to get the relation between each and every variable present in Pandas DataFrame. It works like a seaborn scatter plot but it plot only two variables plot and sns paiplot plot the pairwise plot of multiple features/variable in a grid format.

```python
#pairplot based on category
sns.pairplot(df, hue = 'Category')
```

```python
#pairplot based on segment
sns.pairplot(df, hue = 'Segment')
```

###### Segment Based Analysis

```python
# Number of customers buy products from the Superstore
print(df['Segment'].value_counts())

# Visualise
sns.countplot(x = df['Segment'])
```

```python
# Analysis of Profit, Discount and Sales
data = df.groupby(['Segment'])[['Profit','Discount','Sales']].mean()
print(data)

# Visualise
# Segment data normalised for plotting
df.groupby(['Segment'])[['Profit','Discount','Sales']].mean().plot(kind = 'pie', subplots = True, startangle=90, figsize=(15,10), autopct='%1.1f%%', normalize=True)
plt.figure(figsize=(10,15))
plt.show()
```

###### Category Based Analysis

```python
# Number of products sold from the Superstore
print(df['Category'].value_counts())

# Visualise
sns.countplot(x = df['Category'])
```

```python
# Analysis of Profit
data = df.groupby(['Category'])[['Profit']].mean()
print(data)

# Visualise
category_profit = data.sort_values('Profit', ascending=False)
category_profit[['Profit']].plot(kind = 'bar', figsize = (5,4))
plt.title('Category-based Profit analysis', fontsize = 20)
plt.xlabel('Category', fontsize = 15)
plt.ylabel('Profit', fontsize = 15)
plt.show()
```

Highest profit: Furniture
<br>Lowest profit: Technology

```python
# Analysis of Sales
df.groupby(['Category'])[['Sales']].mean().plot(kind = 'pie', subplots = True, startangle=90, figsize=(15,10), autopct='%1.1f%%')
plt.figure(figsize=(10,15))
plt.show()
```

Maximum Sales in Category: Technology

###### Sub-Category Based Analysis

```python
# Number of sub categories of products sold from the Superstore
print(df['Sub-Category'].value_counts())

# Visualise
plt.figure(figsize = (20,8))
sns.countplot(x = df['Sub-Category'])
plt.title('Sub-Category', fontsize = 20)
```

```python
# Analysis of Profit, Discount and Sales
data = df.groupby(['Sub-Category'])[['Profit','Sales','Discount']].mean()
data
```

```python
# Visualise Profit
subcategory_profit = data.sort_values('Profit', ascending=False)
subcategory_profit[['Profit']].plot(kind = 'bar', figsize = (10,5))
plt.title('Sub-Category-based Profit analysis', fontsize = 20)
plt.xlabel('Sub-Category', fontsize = 15)
plt.ylabel('Profit', fontsize = 15)
plt.show()
```

Highest Profit: Copiers
<br>Lowest Profite: Tables

```python
# Analysis of Discount
data.groupby(['Sub-Category'])[['Discount']].mean().plot(kind = 'pie', subplots = True, startangle=90, figsize=(15,10), autopct='%1.1f%%')
plt.figure(figsize=(10,15))
plt.show()
```

Top 3 High Discount Products: Binders, Machines, Tables

###### State Based Analysis

```python
# State-wise dealings
plt.figure(figsize=(18,5))
sns.countplot(x = df['State'])
plt.xticks(rotation=90)
plt.show()

state = df['State'].value_counts()
print("Average Number of Deals per state is", state.mean())
```

```python
# Analysis of Profit, Discount and Sales
data = df.groupby(['State'])[['Profit','Discount','Sales']].mean()
data.head(10)
```

```python
# Visualise Profit
state_profit = data.sort_values('Profit', ascending=False)
state_profit[['Profit']].plot(kind = 'bar', figsize = (15,10))
plt.title('State-wise Profit analysis', fontsize = 20)
plt.xlabel('States', fontsize = 20)
plt.ylabel('Profit', fontsize = 20)
plt.show()
```

Highest Profit: Vermont
<br>Lowest Profit: Ohio

```python
# Visualise Sales
state_sales = data.sort_values('Sales', ascending=False)
state_sales[['Sales']].plot(kind = 'bar', figsize = (15,10))
plt.title('State-wise Sales analysis', fontsize = 20)
plt.xlabel('States', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.show()
```

Highest amount of Sales: Wyoming

###### City Based Analysis

```python
#City-wise dealings
data = df['City'].value_counts()
data = data.head(25)
data.plot(kind = 'bar', figsize=(18,5))
plt.xticks(rotation=90)
plt.show()
```

```python
# Analysis of Profit
data = df.groupby(['City'])[['Profit']].mean()
data.head(10)
```

```python
city_profit = data.sort_values('Profit', ascending=False).head(10)
city_profit[['Profit']].plot(kind = 'bar',figsize = (10,8))
plt.title('City-wise Profit analysis',fontsize = 20)
plt.xlabel('City', fontsize = 20)
plt.ylabel('Profit',fontsize = 20)
plt.show()
```

```python
city_profit = data.sort_values('Profit', ascending=False).tail(10)
city_profit[['Profit']].plot(kind = 'bar',figsize = (10,8))
plt.title('City-wise Profit analysis',fontsize = 20)
plt.xlabel('City', fontsize = 20)
plt.ylabel('Profit',fontsize = 20)
plt.show()
```

Highest Profit: Jamestown
<br>Lowest Profit: Bethlehem
